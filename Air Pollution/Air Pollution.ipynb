```python
X = df.values[:,:-1]
Y = df.values[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
# X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
lr.fit(X,Y)
```

```python
grad = lr.coef_
```
