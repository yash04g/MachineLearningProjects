```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('./Jamie_Before.jpg')
```

```python
plt.imshow(img1)
plt.axis('off')
plt.show()
```

```python
eye_detector = cv2.CascadeClassifier('./third-party/frontalEyes35x16.xml')
```

```python
eyes = eye_detector.detectMultiScale(img1,1.3,5)
print(eyes)
```

```python
print(eyes.shape[0])
```

```python
for eye in eyes:
    x,y,w,h = eye
    cv2.rectangle(img1,(x+int(w/4),y+int(h/10)),(x+2*w,y+int(h/3)),(0,0,0),2)

plt.figure(figsize = (10,10))
plt.imshow(img1)
plt.show()
```
